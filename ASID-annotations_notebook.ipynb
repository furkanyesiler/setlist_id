Lastly, you can find the concert subsets in the `concert_subsets.json` file.

```python
import json
```

```python
with open('./concert_subsets.json') as f:
    subsets = json.load(f)
```

```python
subsets.keys()
```

```python
for key in subsets.keys():
    print('The concerts under subset {}:'.format(key))
    print(subsets[key])
    print('-----------\n')
```
